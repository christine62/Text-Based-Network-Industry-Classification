```python
map_key = compf.gvkey.apply(str) + "_" + compf.fyear.apply(lambda x: str(int(x)))
compf['map_key'] = map_key
```

```python
g = compf.groupby('gvkey')
avg_at = {}

for key, df in g:
    df = df.sort_values('fyear').reset_index(drop=True)
    for i in range(len(df)):
        if i == 0:
            avg_at[df.ix[i,'map_key']] = df.ix[i,'at'] 
            
        else:
            avg_at[df.ix[i,'map_key']] = (df.ix[i,'at']+df.ix[i-1,'at'])/2
            
```

```python
compf['avg_at'] = compf['map_key'].map(avg_at)
```

```python
#WC = act - che -(lct-dlc)
#NCO = at-act-ivao-(lt-lct-dltt)
#FIN = ivst + ivao -(dltt+dlc+pstk)
#soft_asset = (at-ppent-che)/avg_at
compf['wc']=(compf['act']-compf['che'])-(compf['lct']-compf['dlc'])
compf['nco']=(compf['at']-compf['act']-compf['ivao'])-(compf['lt']-compf['lct']-compf['dltt'])
compf['fin']=(compf['ivst']+compf['ivao'])-(compf['dltt']+compf['dlc']+compf['pstk'])
compf['soft_asset'] = (compf['at']-compf['ppent']-compf['che'])/compf['avg_at']
```

```python
rsst_acc={} #rsst_acc= (d_WC + d_NCO + d_FIN)/avg_at
ch_rec = {}#ch_rec=d_rect/avg_at
ch_inv = {}#ch_inv = d_invt/avg_at
ch_roa = {}#ch_roa= [Earningst (DATA 18) / Average total assetst ] ) -[Earningst-1 /Average total assetst-1] 
cs = {}#ch_cs= sale- d_rect
for key, df in g:
    df = df.sort_values('fyear').reset_index(drop=True)
    for i in range(len(df)):
        if i == 0:
            rsst_acc[df.ix[i,'map_key']] = 0
            ch_rec[df.ix[i,'map_key']] = 0
            ch_inv[df.ix[i,'map_key']] = 0
            cs[df.ix[i,'map_key']] = 0
        else:
            rsst_acc[df.ix[i,'map_key']] = ((df.ix[i,'wc']-df.ix[i-1,'wc'])+(df.ix[i,'nco']-df.ix[i-1,'nco'])+(df.ix[i,'fin']-df.ix[i-1,'fin']))/df.ix[i,'avg_at']
            ch_rec[df.ix[i,'map_key']] = (df.ix[i,'rect']-df.ix[i-1,'rect'])/df.ix[i,'avg_at']
            ch_inv[df.ix[i,'map_key']] = (df.ix[i,'invt']-df.ix[i-1,'invt'])/df.ix[i,'avg_at']
            ch_roa[df.ix[i,'map_key']] = (df.ix[i,'ib']/df.ix[i,'avg_at'])/(df.ix[i-1,'ib']/df.ix[i-1,'avg_at'])
            cs[df.ix[i,'map_key']] = (df.ix[i,'sale']-(df.ix[i,'rect']-df.ix[i-1,'rect']))
```

```python
compf['rsst_acc'] = compf['map_key'].map(rsst_acc)
compf['ch_rec'] = compf['map_key'].map(ch_rec)
compf['ch_inv'] = compf['map_key'].map(ch_inv)
compf['ch_roa'] = compf['map_key'].map(ch_roa)
compf['cs'] = compf['map_key'].map(cs)
```

```python
ch_cs = {}#ch_cs= sale- d_rect
for key, df in g:
    df = df.sort_values('fyear').reset_index(drop=True)
    for i in range(len(df)):
        if i == 0:
            ch_cs[df.ix[i,'map_key']] = 0
        else:
            ch_cs[df.ix[i,'map_key']] = (df.ix[i,'cs']-df.ix[i-1,'cs'])/df.ix[i-1,'cs']
```

```python
compf['ch_cs'] = compf['map_key'].map(ch_cs)
```

```python
#issue=An indicator variable coded 1 if the firm issued securities during year t (i.e., an indicator variable coded 1 if DATA 108(sstk) > 0 or DATA111(dltis) > 0)  
issue = {}
for key, df in g:
    df = df.sort_values('fyear').reset_index(drop=True)
    for i in range(len(df)):
        if (df.ix[i,'sstk']>0) or (df.ix[i,'dltis']>0):
            issue[df.ix[i,'map_key']] = 1 
        else:
            issue[df.ix[i,'map_key']] = 0
```

```python
compf['issue'] = compf['map_key'].map(issue)
```
